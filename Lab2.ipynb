```python
class Employee:
    def __init__(self, name, l_name, salary, exp):
        self.name = name
        self.last_name = l_name
        self.base_salary = salary
        self.experience = exp


class Developer(Employee):
    def __init__(self, name, l_name, salary, exp):
       super().__init__(name, l_name, salary, exp)
       

class Designer(Employee):
    def __init__(self, name, l_name, salary, exp, er):
       super().__init__(name, l_name, salary, exp)
       self.er = er

class Manager(Employee):
    def __init__(self, name, l_name, salary, exp, team):
       super().__init__(name, l_name, salary, exp)
       self.team = team   


class Department:
    def __init__(self, managers):
         self.managers = managers
    
    def giveSalary(self):
        print('\n'.join(self.countedSalary()))

    def countedSalary(self):
        ans = []
        for m in self.managers:
            salary = m.base_salary
            if m.experience > 5:   salary = salary*1.2 + 500
            elif m.experience > 2: salary += 200

            if len(m.team)>10: salary += 300
            elif len(m.team)>5: salary += 200
            
            if sum(isinstance(x, Developer) for x in m.team)>len(m.team)/2: salary *= 1.1
            ans.append(f'\n{m.name} {m.last_name} отримав(-ла) {salary} Укробаксів')

            for e in m.team:
                salary = e.base_salary

                if e.experience > 5:
                    salary = salary*1.2 + 500
                elif e.experience > 2:
                    salary += 200
                
                if isinstance(e, Designer):
                    salary *= e.er

                ans.append(f'   {e.name} {e.last_name} отримав(-ла) {salary} Укробаксів')

        return ans
department = Department(
    [
        Manager('Володимир', 'Ємченко', 3075, 7,
                [
                    Designer('Анісій',  'Вірченко', 4600, 4,   1),
                    Designer('Ладолюб',  'Шестопал',      4110, 1.5, 0.6),
                    Designer('Фауст',   'Кравців',     3800, 3.5, 0.4),
                    Designer('Іван',    'Шимків',  2500, 1.5, 0.2),
                    Developer('Юрій', 'Гойда',   4800, 5)
                ]),
        Manager('Юрій', 'Маляренко', 4300, 6,
                [
                    Designer('Владислав', 'Кологойда',    3900,   1,   0.6),
                    Designer('Йозеф',    'Ляшко', 2700,   0.5, 0.4),
                    Developer('Едуард',    'Ковалюк',  4200,   3),
                    Developer('Руслан',  'Погрібний',    4970,   4),
                    Developer('Ждана',  'Котовська',   2700, 2),
                ]),
        Manager('Світлана', 'Мірошниченко', 2600,5,
                [
                    Designer('Анастасія',     'Кравчук',    4250, 1,   0.7),
                    Designer('ОКсана',    'Олексієнко', 3600, 1.5, 0.3),
                    Designer('Тарас',    'Шевченко',    2600, 1,   0.1),
                    Designer('Іван',    'Франко',    3600, 3,   0.4),
                    Designer('Леся',    'Українка',  2300, 1.5, 0.3),
                    Designer('Павло',    'Зібров',    3800, 1.5, 0.3),
                    Developer('Михайло',   'Поплавський',    3900, 2.5),
                    Developer('Єгор',    'Ззагор',     3200, 2),
                    Developer('Надія','Павлюк',   1900, 1),
                    Developer('Петро',    'Порошенко',   1200, 0.5),
                    Developer('Олексій',  'Арестович',    700, 1)
                ])
        
    ])
department.giveSalary()
```
